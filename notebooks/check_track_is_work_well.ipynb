# 개요

해당 노트북 파일은, 수집한 데이터가 제대로 잘 수집되었는제

```python
from trailine.graph_map import GraphMap

g: GraphMap = GraphMap()
```

```python
# 데이터 세팅
## 설악산 오색 쪽 데이터만 가져오기
## 오색코스 루트: 남설악탐방지원센터 - 오색1쉼터 - 설악06-06 - 대청봉

# 웨이포인트 가져오기
waypoints = g.search_waypoints(parent_place_name="설악산")
waypoint_names = ("남설악탐방지원센터", "오색1쉼터", "설약06-06", "대청봉")
waypoint_codes = {
    w.name: g.search_waypoints(parent_place_name="설악산", name=w.name)[0].code
     for w in waypoints
}

# 트랙(간선) 가져오기
# 0: 남설악탐방지원센터 -> 오색1쉼터
# 1: 오색1쉽터 -> 설악06-06
# 2: 설악06-06 -> 대청봉

tracks = [
    g.tracks[waypoint_codes["남설악탐방지원센터"]][waypoint_codes["오색1쉼터"]][0],
    g.tracks[waypoint_codes["오색1쉼터"]][waypoint_codes["설악06-06"]][0],
    g.tracks[waypoint_codes["설악06-06"]][waypoint_codes["대청봉"]][0]
]
```

```python
from trailine.vars import WAYPOINT_LIST_PATH

WAYPOINT_LIST_PATH
```

```python
# 지도에서 제대로 트랙이 수집되었는지 확인

import folium
from statistics import mean

track = tracks[0]
all_track_circuit = []
for track in tracks:
    all_track_circuit += track.circuit

mean_lat = mean([p.lat for p in all_track_circuit])
mean_lon = mean([p.lon for p in all_track_circuit])

m = folium.Map(location=[mean_lat, mean_lon], zoom_start=14)

for track in tracks:
    folium.PolyLine([(p.lat, p.lon) for p in track.circuit], color="red", weight=5).add_to(m)


m
```
